##  Definiciones y Teoremas de Pruebas de Hip贸tesis

### Definici贸n

Una **prueba de hip贸tesis** es un procedimiento estad铆stico para decidir entre dos afirmaciones contrapuestas sobre un par谩metro poblacional a partir de datos muestrales.

Se plantea:

- **Hip贸tesis nula** $H_0$
- **Hip贸tesis alternativa** $H_1$

El objetivo es evaluar si los datos observados proporcionan suficiente evidencia contra $H_0$.

### Pasos en una prueba de hip贸tesis:

1. Formular $H_0$ y $H_1$
2. Elegir el nivel de significancia $\alpha$
3. Determinar la estad铆stica de prueba
4. Definir la regi贸n cr铆tica
5. Tomar decisi贸n con base en los datos

### Tipos de errores:

- **Error tipo I**: Rechazar $H_0$ cuando es verdadera.
- **Error tipo II**: No rechazar $H_0$ cuando es falsa.

### Estad铆sticas comunes:

Para una media poblacional:

\begin{equation}
Z = \frac{\bar{X} - \mu_0}{\sigma / \sqrt{n}}
\end{equation}

Para una proporci贸n poblacional:

\begin{equation}
Z = \frac{\hat{p} - p_0}{\sqrt{ \frac{p_0 (1-p_0)}{n} }}
\end{equation}

Para una media con varianza desconocida (Distribuci贸n t):

\begin{equation}
t = \frac{\bar{X} - \mu_0}{S / \sqrt{n}}
\end{equation}

### Propiedades importantes

- Si $\alpha$ disminuye, $\beta$ aumenta (y viceversa)
- El valor $p$ representa la probabilidad de obtener un valor tan extremo como el observado si $H_0$ fuera cierta



```python

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Par谩metros
mu_0 = 50
mu_real = 52
sigma = 5
n = 30
alpha = 0.05

# Simulaci贸n de muestra
np.random.seed(0)
muestra = np.random.normal(mu_real, sigma, n)

# Estad铆stica de prueba
x_bar = np.mean(muestra)
z = (x_bar - mu_0) / (sigma / np.sqrt(n))

# Valor p
p_value = 1 - stats.norm.cdf(z)

# Resultados
print(f"Media muestral: {x_bar:.2f}")
print(f"Estad铆stico Z: {z:.2f}")
print(f"Valor p: {p_value:.4f}")

# Gr谩fica
x = np.linspace(mu_0 - 4*sigma, mu_0 + 4*sigma, 1000)
y = stats.norm.pdf(x, mu_0, sigma/np.sqrt(n))

plt.plot(x, y, label='Distribuci贸n bajo H0')
plt.axvline(x_bar, color='red', linestyle='--', label=f'Media muestra ({x_bar:.2f})')
plt.fill_between(x, 0, y, where=(x >= mu_0 + stats.norm.ppf(1-alpha)*sigma/np.sqrt(n)), color='orange', alpha=0.5, label='Regi贸n de rechazo')
plt.title('Prueba de Hip贸tesis para una Media')
plt.legend()
plt.grid(True)
plt.show()
```
